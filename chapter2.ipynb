```python
#Procedure to generate random samples according to a normal distribution in d dimensions
def gaussian_samples(mean, covariance,num_samples):
    return np.random.multivariate_normal(mean,covariance,num_samples)
```

```python
#calculate discriminant function for a given normal distribution and priort probability
def discriminant(x,d,mean,cov,prior):
    inv_cov = np.linalg.inv(cov)
    determinant = np.linalg.det(cov)

    return -0.5*np.dot((x-mean),inv_cov@(x-mean))-0.5*d*np.log(2*np.pi)-0.5*np.log(determinant)+ np.log(prior)
```
