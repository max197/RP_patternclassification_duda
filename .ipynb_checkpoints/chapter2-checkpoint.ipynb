```python
#Procedure to generate random samples according to a normal distribution in d dimensions
def gaussian_samples(mean, covariance,num_samples):
    return np.random.multivariate_normal(mean,covariance,num_samples)
```

```python
cov = np.array([[6,-3],[-3,3.5]])
mean = np.array((1,2))
x = np.random.rand(2)
np.dot((x-mean),cov@(x-mean))
```

```python
#calculate discriminant function for a given multivariate normal distribution and prior probability
def discriminant(x,d,mean,cov,prior):
    if d==1:
        return -0.5*((x-mean)/cov)**2 - 0.5*np.log(2*np.pi*var) + np.log(prior)
    elif d>1:
        inv_cov = np.linalg.inv(cov)
        determinant = np.linalg.det(cov)
        return -0.5*np.dot((x-mean),inv_cov@(x-mean))-0.5*d*np.log(2*np.pi)-0.5*np.log(np.abs(determinant))+ np.log(prior)
    else:
        raise Exception("dimension must be positive")
```

```python
samples = gaussian_samples(np.array([5,2]),np.array([[1,0],[0,1]]),1000)
samples[:3]
```

```python
plt.plot(samples[:, 0], samples[:, 1], '.', alpha=0.5)
plt.axis('equal')
plt.grid()
plt.show()
```

```python
#dichotomizer 
def dichotomizer(x,mu1,mu2,var1,var2,prior1,prior2):
    g1 = discriminant(x=x,d=1,mean=mu1,cov=var1,prior=prior1)
    g2 = discriminant(x=x,d=1,mean=mu2,cov=var2,prior=prior2)
    print(g1-g2)
    if  g1-g2 >0:
        return "omega_1"
    else:
        return "omega_2"
        
```

```python
mu1 = np.mean(x1w1)
var1 = np.var(x1w1)
mu2 = np.mean(x1w2)
var2 = np.var(x1w2)
prior1=prior2=0.5
for x in x1w2:
    print(dichotomizer(x,mu1,mu2,var1,var2,prior1,prior2))
```
